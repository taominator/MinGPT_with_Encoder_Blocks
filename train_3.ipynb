```python
import torch
from torch.utils.data import Dataset
from torch.utils.data.dataloader import DataLoader
from mingpt.utils import set_seed
set_seed(3407)
```

```python
from mingpt.chunkedtextdataset import ChunkedTextDataset

# print an example instance of the dataset
train_dataset = ChunkedTextDataset('99')
# Fetch and print the first 20 elements
for idx, (x, y) in enumerate(train_dataset):
    if idx >= 20:
        break
    print(f"Data {idx + 1} - Input: {x} (TYPE: {type(x)}), Target: {y} (TYPE: {type(y)})")
```

```python
# create a GPT instance
from mingpt.model import GPT

model_config = GPT.get_default_config()
model_config.model_type = None
model_config.n_layer = 12
model_config.n_head = 16
model_config.n_embd = 1024
model_config.n_encoder_layer = 12
model_config.vocab_size = 50257
model_config.block_size = 1024

#type_given = (model_config.model_type is not None) & (model_config.n_encoder_layer is not None)
#print(type_given)
#params_given = all([model_config.n_layer is not None, model_config.n_head is not None, model_config.n_embd is not None, model_config.n_encoder_layer is not None])
#print(params_given)
#print(type_given ^ params_given)


model = GPT(model_config)
```

```python
# create a Trainer object
from mingpt.trainer import Trainer

train_config = Trainer.get_default_config()
train_config.learning_rate = 5e-4 # the model we're using is so small that we can go a bit faster
trainer = Trainer(train_config, model, train_dataset)
```

```python
trainer.run()
```
